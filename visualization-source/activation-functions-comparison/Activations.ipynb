```python
from IPython.display import HTML, display

display(HTML(r"""
<div id="act_fix_root">
  <style>
    #act_fix_root * { box-sizing: border-box; }
    #act_fix_root {
      font-family: Inter, system-ui, -apple-system, BlinkMacSystemFont, "Segoe UI", sans-serif;
      color: #182234;
      max-width: 1380px;
      margin: 18px auto;
      --bg: #f4f8ff;
      --panel: rgba(255,255,255,0.88);
      --line: rgba(15,23,42,0.08);
      --text: #182234;
      --muted: #6b7280;
      --blue: #2563eb;
      --green: #2f7d4f;
      --orange: #b86a1a;
      --red: #ba4141;
      --shadow: 0 18px 44px rgba(35,64,124,0.08);
    }

    #act_fix_root .app {
      background:
        radial-gradient(circle at top right, rgba(37,99,235,0.08), transparent 26%),
        linear-gradient(180deg, #fbfdff 0%, #f4f8ff 100%);
      border: 1px solid rgba(15,23,42,0.05);
      border-radius: 28px;
      overflow: hidden;
      box-shadow: var(--shadow);
    }

    #act_fix_root .hero {
      padding: 28px 30px 18px;
      border-bottom: 1px solid rgba(15,23,42,0.05);
      background: rgba(255,255,255,0.58);
    }

    #act_fix_root .eyebrow {
      font-size: 11px;
      letter-spacing: 0.12em;
      text-transform: uppercase;
      color: #5d79a8;
      font-weight: 700;
      margin-bottom: 10px;
    }

    #act_fix_root .title {
      font-size: clamp(28px, 4vw, 42px);
      font-weight: 760;
      line-height: 1.02;
      letter-spacing: -0.04em;
      margin-bottom: 10px;
    }

    #act_fix_root .title span {
      background: linear-gradient(135deg, #1d4ed8, #60a5fa);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
    }

    #act_fix_root .desc {
      font-size: 15px;
      color: var(--muted);
      line-height: 1.72;
      max-width: 820px;
    }

    #act_fix_root .content {
      display: grid;
      grid-template-columns: minmax(0, 1.18fr) minmax(330px, 0.82fr);
      gap: 16px;
      padding: 18px;
    }

    #act_fix_root .panel {
      background: var(--panel);
      border: 1px solid rgba(15,23,42,0.05);
      border-radius: 22px;
      box-shadow: var(--shadow);
    }

    #act_fix_root .left {
      display: grid;
      grid-template-rows: auto auto;
      gap: 16px;
    }

    #act_fix_root .main-panel,
    #act_fix_root .bottom-panel,
    #act_fix_root .side {
      padding: 18px;
    }

    #act_fix_root .head {
      display: flex;
      justify-content: space-between;
      align-items: flex-start;
      gap: 12px;
      margin-bottom: 12px;
    }

    #act_fix_root .panel-title {
      font-size: 18px;
      font-weight: 700;
      letter-spacing: -0.02em;
      margin-bottom: 3px;
    }

    #act_fix_root .panel-sub {
      font-size: 13px;
      color: var(--muted);
      line-height: 1.6;
    }

    #act_fix_root .badge {
      display: inline-flex;
      align-items: center;
      gap: 6px;
      padding: 6px 10px;
      border-radius: 999px;
      background: #eaf1ff;
      color: var(--blue);
      font-size: 12px;
      font-weight: 700;
      white-space: nowrap;
    }

    #act_fix_root .canvas-wrap {
      width: 100%;
      border-radius: 18px;
      overflow: hidden;
      border: 1px solid rgba(15,23,42,0.05);
      background: linear-gradient(180deg, rgba(255,255,255,0.94), rgba(247,250,255,0.96));
    }

    #act_fix_root .canvas-big {
      height: 430px;
    }

    #act_fix_root .canvas-small {
      height: 260px;
    }

    #act_fix_root canvas {
      display: block;
      width: 100%;
      height: 100%;
    }

    #act_fix_root .legend {
      display: flex;
      gap: 12px;
      flex-wrap: wrap;
      margin-top: 12px;
      color: var(--muted);
      font-size: 12px;
    }

    #act_fix_root .legend-item {
      display: inline-flex;
      align-items: center;
      gap: 8px;
    }

    #act_fix_root .legend-line {
      width: 18px;
      height: 0;
      border-top: 2px solid var(--blue);
    }

    #act_fix_root .legend-line.orange {
      border-top-color: var(--orange);
    }

    #act_fix_root .legend-dot {
      width: 10px;
      height: 10px;
      border-radius: 50%;
      background: var(--blue);
    }

    #act_fix_root .legend-dot.orange {
      background: var(--orange);
    }

    #act_fix_root .side {
      display: flex;
      flex-direction: column;
      gap: 16px;
    }

    #act_fix_root .card {
      background: rgba(255,255,255,0.95);
      border: 1px solid rgba(15,23,42,0.05);
      border-radius: 16px;
      padding: 15px 16px;
    }

    #act_fix_root .label {
      font-size: 12px;
      font-weight: 700;
      margin-bottom: 10px;
    }

    #act_fix_root .btn-row {
      display: flex;
      gap: 8px;
      flex-wrap: wrap;
    }

    #act_fix_root .btn {
      border: 1px solid rgba(15,23,42,0.08);
      background: rgba(255,255,255,0.86);
      color: var(--text);
      border-radius: 999px;
      padding: 9px 14px;
      font-size: 12px;
      font-weight: 700;
      cursor: pointer;
      transition: 0.16s ease;
      font-family: inherit;
    }

    #act_fix_root .btn:hover {
      transform: translateY(-1px);
      box-shadow: 0 10px 18px rgba(15,23,42,0.08);
    }

    #act_fix_root .btn.active {
      background: var(--blue);
      border-color: var(--blue);
      color: #fff;
    }

    #act_fix_root .slider-top {
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 10px;
      margin-bottom: 8px;
    }

    #act_fix_root .slider-title {
      font-size: 13px;
      font-weight: 700;
    }

    #act_fix_root .pill {
      min-width: 76px;
      text-align: center;
      padding: 6px 10px;
      border-radius: 999px;
      background: #f1f5f9;
      color: #475569;
      font-size: 12px;
      font-weight: 700;
      font-variant-numeric: tabular-nums;
    }

    #act_fix_root input[type="range"] {
      width: 100%;
      accent-color: var(--blue);
      cursor: pointer;
      margin: 6px 0 2px;
    }

    #act_fix_root .subtle {
      font-size: 12px;
      line-height: 1.6;
      color: var(--muted);
    }

    #act_fix_root .status {
      display: inline-flex;
      align-items: center;
      gap: 8px;
      padding: 8px 12px;
      border-radius: 999px;
      font-size: 12px;
      font-weight: 700;
      margin-bottom: 10px;
      border: 1px solid transparent;
    }

    #act_fix_root .status-dot {
      width: 8px;
      height: 8px;
      border-radius: 50%;
    }

    #act_fix_root .metrics {
      display: grid;
      grid-template-columns: 1fr 1fr;
      gap: 10px;
    }

    #act_fix_root .metric {
      background: #f8fbff;
      border: 1px solid rgba(15,23,42,0.05);
      border-radius: 16px;
      padding: 14px;
    }

    #act_fix_root .metric-k {
      font-size: 11px;
      color: #7b8797;
      margin-bottom: 6px;
    }

    #act_fix_root .metric-v {
      font-size: 24px;
      font-weight: 760;
      letter-spacing: -0.03em;
      font-variant-numeric: tabular-nums;
    }

    #act_fix_root .metric-note {
      margin-top: 8px;
      font-size: 11px;
      color: var(--muted);
      line-height: 1.5;
    }

    #act_fix_root .formula {
      font-size: 13px;
      line-height: 1.8;
      color: var(--text);
      white-space: pre-line;
    }

    #act_fix_root .formula strong {
      color: var(--blue);
    }

    #act_fix_root .explain-title {
      font-size: 14px;
      font-weight: 700;
      margin-bottom: 8px;
    }

    #act_fix_root .explain-text {
      font-size: 13px;
      color: var(--muted);
      line-height: 1.7;
    }

    @media (max-width: 1100px) {
      #act_fix_root .content {
        grid-template-columns: 1fr;
      }
    }

    @media (max-width: 700px) {
      #act_fix_root .metrics {
        grid-template-columns: 1fr;
      }
    }
  </style>

  <div class="app">
    <div class="hero">
      <div class="eyebrow">INTERACTIVE ACTIVATION EXPLORER</div>
      <div class="title">Activation functions with a <span>working chart</span></div>
      <div class="desc">
        This version avoids the broken rendering issue. Move the input z, switch activations, or press play.
        You will see both the function curve and the derivative curve update together.
      </div>
    </div>

    <div class="content">
      <div class="left">
        <div class="panel main-panel">
          <div class="head">
            <div>
              <div class="panel-title">Function response</div>
              <div class="panel-sub">The blue point moves on the activation curve as the input changes.</div>
            </div>
            <div class="badge" id="fix_badge">ReLU</div>
          </div>
          <div class="canvas-wrap canvas-big">
            <canvas id="fix_main_canvas"></canvas>
          </div>
          <div class="legend">
            <div class="legend-item"><span class="legend-line"></span> activation curve</div>
            <div class="legend-item"><span class="legend-dot"></span> current input-output point</div>
          </div>
        </div>

        <div class="panel bottom-panel">
          <div class="head">
            <div>
              <div class="panel-title">Derivative response</div>
              <div class="panel-sub">This is what really explains saturation, dead neurons, and gradient flow.</div>
            </div>
            <div class="badge" id="fix_region_badge">Active region</div>
          </div>
          <div class="canvas-wrap canvas-small">
            <canvas id="fix_deriv_canvas"></canvas>
          </div>
          <div class="legend">
            <div class="legend-item"><span class="legend-line orange"></span> derivative curve</div>
            <div class="legend-item"><span class="legend-dot orange"></span> current derivative point</div>
          </div>
        </div>
      </div>

      <div class="panel side">
        <div class="card">
          <div class="label">Choose activation</div>
          <div class="btn-row" id="fix_act_buttons">
            <button class="btn active" data-act="relu">ReLU</button>
            <button class="btn" data-act="leaky">Leaky ReLU</button>
            <button class="btn" data-act="sigmoid">Sigmoid</button>
            <button class="btn" data-act="tanh">Tanh</button>
          </div>
        </div>

        <div class="card">
          <div class="slider-top">
            <div class="slider-title">Input value z</div>
            <div class="pill" id="fix_z_val">1.20</div>
          </div>
          <input type="range" id="fix_z_slider" min="-6" max="6" step="0.01" value="1.2">
          <div class="subtle">Drag this to see how the output and the local slope change together.</div>
        </div>

        <div class="card">
          <div class="label">Playback</div>
          <div class="btn-row">
            <button class="btn" id="fix_play">Play</button>
            <button class="btn" id="fix_pause">Pause</button>
            <button class="btn" id="fix_reset">Reset</button>
          </div>
        </div>

        <div>
          <div class="status" id="fix_status">
            <span class="status-dot" id="fix_status_dot"></span>
            <span id="fix_status_text">Active region</span>
          </div>

          <div class="metrics">
            <div class="metric">
              <div class="metric-k">Current output</div>
              <div class="metric-v" id="fix_out_val">1.20</div>
              <div class="metric-note">a = g(z)</div>
            </div>
            <div class="metric">
              <div class="metric-k">Current derivative</div>
              <div class="metric-v" id="fix_df_val">1.00</div>
              <div class="metric-note">g'(z)</div>
            </div>
          </div>
        </div>

        <div class="card formula" id="fix_formula">
<strong>Function</strong>
g(z) = max(0, z)

<strong>Derivative</strong>
g'(z) = 1 for z > 0, 0 for z < 0
        </div>

        <div class="card">
          <div class="explain-title" id="fix_explain_title">Why ReLU is common</div>
          <div class="explain-text" id="fix_explain_text">
            On the positive side, the slope stays strong, so gradients do not shrink quickly.
            On the negative side, the neuron outputs zero and stops passing gradient.
          </div>
        </div>
      </div>
    </div>
  </div>

  <script>
    (function() {
      const mainCanvas = document.getElementById('fix_main_canvas');
      const derivCanvas = document.getElementById('fix_deriv_canvas');
      const zSlider = document.getElementById('fix_z_slider');
      const zVal = document.getElementById('fix_z_val');
      const outVal = document.getElementById('fix_out_val');
      const dfVal = document.getElementById('fix_df_val');
      const badge = document.getElementById('fix_badge');
      const regionBadge = document.getElementById('fix_region_badge');
      const status = document.getElementById('fix_status');
      const statusDot = document.getElementById('fix_status_dot');
      const statusText = document.getElementById('fix_status_text');
      const formula = document.getElementById('fix_formula');
      const explainTitle = document.getElementById('fix_explain_title');
      const explainText = document.getElementById('fix_explain_text');
      const playBtn = document.getElementById('fix_play');
      const pauseBtn = document.getElementById('fix_pause');
      const resetBtn = document.getElementById('fix_reset');

      const state = {
        activation: 'relu',
        z: 1.2,
        playing: false,
        dir: 1,
        alpha: 0.1,
        raf: null
      };

      const META = {
        relu: {
          name: 'ReLU',
          formula: "g(z) = max(0, z)",
          derivative: "g'(z) = 1 for z > 0, 0 for z < 0",
          title: "Why ReLU is common",
          text: "On the positive side, the slope stays strong, so gradients do not shrink quickly. On the negative side, the neuron outputs zero and stops passing gradient."
        },
        leaky: {
          name: 'Leaky ReLU',
          formula: "g(z) = max(0.1z, z)",
          derivative: "g'(z) = 1 for z > 0, 0.1 for z < 0",
          title: "Why Leaky ReLU is safer",
          text: "It keeps a small slope on the negative side, so gradients do not completely die there."
        },
        sigmoid: {
          name: 'Sigmoid',
          formula: "g(z) = 1 / (1 + e^(-z))",
          derivative: "g'(z) = g(z)(1 - g(z))",
          title: "Why sigmoid can saturate",
          text: "Far on the left or right, the curve becomes flat. Then the derivative becomes very small, so gradients weaken."
        },
        tanh: {
          name: 'Tanh',
          formula: "g(z) = tanh(z)",
          derivative: "g'(z) = 1 - tanh(z)^2",
          title: "Why tanh is nicer than sigmoid",
          text: "It is zero-centered, which is often nicer for optimization, but it still saturates at both ends."
        }
      };

      function f(z) {
        if (state.activation === 'relu') return Math.max(0, z);
        if (state.activation === 'leaky') return z >= 0 ? z : state.alpha * z;
        if (state.activation === 'sigmoid') return 1 / (1 + Math.exp(-z));
        if (state.activation === 'tanh') return Math.tanh(z);
        return z;
      }

      function df(z) {
        if (state.activation === 'relu') return z > 0 ? 1 : 0;
        if (state.activation === 'leaky') return z > 0 ? 1 : state.alpha;
        if (state.activation === 'sigmoid') {
          const s = 1 / (1 + Math.exp(-z));
          return s * (1 - s);
        }
        if (state.activation === 'tanh') {
          const t = Math.tanh(z);
          return 1 - t * t;
        }
        return 1;
      }

      function fmt(x) {
        const s = Number(x).toFixed(2);
        return s === "-0.00" ? "0.00" : s;
      }

      function regionInfo(z) {
        if (state.activation === 'relu') {
          if (z < 0) return { label: "Dead zone", short: "Dead zone", color: "#64748b", bg: "rgba(100,116,139,0.10)" };
          if (Math.abs(z) < 0.15) return { label: "Boundary region", short: "Boundary", color: "#b86a1a", bg: "rgba(184,106,26,0.10)" };
          return { label: "Active region", short: "Active region", color: "#2f7d4f", bg: "rgba(47,125,79,0.10)" };
        }
        if (state.activation === 'leaky') {
          if (z < 0) return { label: "Weak negative-slope region", short: "Weak slope", color: "#b86a1a", bg: "rgba(184,106,26,0.10)" };
          return { label: "Active region", short: "Active region", color: "#2f7d4f", bg: "rgba(47,125,79,0.10)" };
        }
        if (state.activation === 'sigmoid') {
          if (z < -3 || z > 3) return { label: "Saturated region", short: "Saturation", color: "#ba4141", bg: "rgba(186,65,65,0.10)" };
          return { label: "Sensitive middle region", short: "Sensitive", color: "#2f7d4f", bg: "rgba(47,125,79,0.10)" };
        }
        if (state.activation === 'tanh') {
          if (z < -2.5 || z > 2.5) return { label: "Saturated region", short: "Saturation", color: "#ba4141", bg: "rgba(186,65,65,0.10)" };
          return { label: "Responsive middle region", short: "Responsive", color: "#2f7d4f", bg: "rgba(47,125,79,0.10)" };
        }
        return { label: "Region", short: "Region", color: "#2f7d4f", bg: "rgba(47,125,79,0.10)" };
      }

      function resizeCanvas(canvas) {
        const dpr = window.devicePixelRatio || 1;
        const rect = canvas.getBoundingClientRect();
        canvas.width = Math.round(rect.width * dpr);
        canvas.height = Math.round(rect.height * dpr);
        const ctx = canvas.getContext('2d');
        ctx.setTransform(dpr, 0, 0, dpr, 0, 0);
        return { ctx, w: rect.width, h: rect.height };
      }

      function drawAxes(ctx, w, h, xMin, xMax, yMin, yMax, xTicks, yTicks, area) {
        const { ml, mr, mt, mb } = area;
        const pw = w - ml - mr;
        const ph = h - mt - mb;

        const xMap = x => ml + ((x - xMin) / (xMax - xMin)) * pw;
        const yMap = y => mt + (1 - (y - yMin) / (yMax - yMin)) * ph;

        ctx.strokeStyle = "rgba(15,23,42,0.06)";
        ctx.lineWidth = 1;

        xTicks.forEach(t => {
          const x = xMap(t);
          ctx.beginPath();
          ctx.moveTo(x, mt);
          ctx.lineTo(x, mt + ph);
          ctx.stroke();
        });

        yTicks.forEach(t => {
          const y = yMap(t);
          ctx.beginPath();
          ctx.moveTo(ml, y);
          ctx.lineTo(ml + pw, y);
          ctx.stroke();
        });

        ctx.strokeStyle = "rgba(15,23,42,0.26)";
        ctx.lineWidth = 1.4;

        ctx.beginPath();
        ctx.moveTo(ml, yMap(0));
        ctx.lineTo(ml + pw, yMap(0));
        ctx.stroke();

        ctx.beginPath();
        ctx.moveTo(xMap(0), mt);
        ctx.lineTo(xMap(0), mt + ph);
        ctx.stroke();

        ctx.fillStyle = "#8793a5";
        ctx.font = "11px Inter, sans-serif";
        ctx.textAlign = "center";
        xTicks.forEach(t => {
          ctx.fillText(String(t), xMap(t), h - 10);
        });

        ctx.textAlign = "right";
        yTicks.forEach(t => {
          ctx.fillText(String(t), ml - 10, yMap(t) + 4);
        });

        return { xMap, yMap };
      }

      function drawRegionBands(ctx, xMap, yTop, yBottom) {
        ctx.save();
        if (state.activation === 'relu') {
          ctx.fillStyle = "rgba(100,116,139,0.08)";
          ctx.fillRect(xMap(-6), yTop, xMap(0) - xMap(-6), yBottom - yTop);
          ctx.fillStyle = "rgba(47,125,79,0.06)";
          ctx.fillRect(xMap(0), yTop, xMap(6) - xMap(0), yBottom - yTop);
        } else if (state.activation === 'leaky') {
          ctx.fillStyle = "rgba(184,106,26,0.08)";
          ctx.fillRect(xMap(-6), yTop, xMap(0) - xMap(-6), yBottom - yTop);
          ctx.fillStyle = "rgba(47,125,79,0.06)";
          ctx.fillRect(xMap(0), yTop, xMap(6) - xMap(0), yBottom - yTop);
        } else if (state.activation === 'sigmoid') {
          ctx.fillStyle = "rgba(186,65,65,0.08)";
          ctx.fillRect(xMap(-6), yTop, xMap(-3) - xMap(-6), yBottom - yTop);
          ctx.fillStyle = "rgba(47,125,79,0.06)";
          ctx.fillRect(xMap(-3), yTop, xMap(3) - xMap(-3), yBottom - yTop);
          ctx.fillStyle = "rgba(186,65,65,0.08)";
          ctx.fillRect(xMap(3), yTop, xMap(6) - xMap(3), yBottom - yTop);
        } else if (state.activation === 'tanh') {
          ctx.fillStyle = "rgba(186,65,65,0.08)";
          ctx.fillRect(xMap(-6), yTop, xMap(-2.5) - xMap(-6), yBottom - yTop);
          ctx.fillStyle = "rgba(47,125,79,0.06)";
          ctx.fillRect(xMap(-2.5), yTop, xMap(2.5) - xMap(-2.5), yBottom - yTop);
          ctx.fillStyle = "rgba(186,65,65,0.08)";
          ctx.fillRect(xMap(2.5), yTop, xMap(6) - xMap(2.5), yBottom - yTop);
        }
        ctx.restore();
      }

      function drawFunctionChart() {
        const { ctx, w, h } = resizeCanvas(mainCanvas);
        ctx.clearRect(0, 0, w, h);

        const area = { ml: 56, mr: 24, mt: 18, mb: 36 };
        const xMin = -6, xMax = 6;

        let yMin = -1.2, yMax = 6.2, yTicks = [-1,0,1,2,3,4,5,6];
        if (state.activation === 'sigmoid') { yMin = -0.1; yMax = 1.1; yTicks = [0,0.25,0.5,0.75,1]; }
        if (state.activation === 'tanh') { yMin = -1.2; yMax = 1.2; yTicks = [-1,-0.5,0,0.5,1]; }

        const maps = drawAxes(ctx, w, h, xMin, xMax, yMin, yMax, [-6,-4,-2,0,2,4,6], yTicks, area);
        const { xMap, yMap } = maps;
        drawRegionBands(ctx, xMap, area.mt, h - area.mb);

        ctx.strokeStyle = "#2563eb";
        ctx.lineWidth = 3;
        ctx.beginPath();
        for (let i = 0; i <= 320; i++) {
          const x = xMin + (xMax - xMin) * i / 320;
          const y = f(x);
          const px = xMap(x);
          const py = yMap(y);
          if (i === 0) ctx.moveTo(px, py);
          else ctx.lineTo(px, py);
        }
        ctx.stroke();

        const cx = xMap(state.z);
        const cy = yMap(f(state.z));

        ctx.setLineDash([5, 6]);
        ctx.strokeStyle = "rgba(37,99,235,0.22)";
        ctx.lineWidth = 1.6;
        ctx.beginPath();
        ctx.moveTo(cx, yMap(yMin));
        ctx.lineTo(cx, cy);
        ctx.stroke();
        ctx.beginPath();
        ctx.moveTo(xMap(0), cy);
        ctx.lineTo(cx, cy);
        ctx.stroke();
        ctx.setLineDash([]);

        const slope = df(state.z);
        const half = 1.1;
        const x1 = state.z - half;
        const x2 = state.z + half;
        const y1 = f(state.z) - slope * half;
        const y2 = f(state.z) + slope * half;

        ctx.strokeStyle = "rgba(15,118,110,0.82)";
        ctx.lineWidth = 2;
        ctx.beginPath();
        ctx.moveTo(xMap(x1), yMap(y1));
        ctx.lineTo(xMap(x2), yMap(y2));
        ctx.stroke();

        ctx.fillStyle = "#ffffff";
        ctx.strokeStyle = "#2563eb";
        ctx.lineWidth = 3;
        ctx.beginPath();
        ctx.arc(cx, cy, 6.5, 0, Math.PI * 2);
        ctx.fill();
        ctx.stroke();

        ctx.fillStyle = "#2563eb";
        ctx.font = "700 12px Inter, sans-serif";
        ctx.textAlign = "left";
        ctx.fillText(`(${fmt(state.z)}, ${fmt(f(state.z))})`, cx + 12, cy - 10);
      }

      function drawDerivativeChart() {
        const { ctx, w, h } = resizeCanvas(derivCanvas);
        ctx.clearRect(0, 0, w, h);

        const area = { ml: 56, mr: 24, mt: 16, mb: 34 };
        const xMin = -6, xMax = 6;

        let yMin = -0.02, yMax = 1.1, yTicks = [0,0.5,1];
        if (state.activation === 'sigmoid') { yMin = -0.02; yMax = 0.32; yTicks = [0,0.1,0.2,0.3]; }
        if (state.activation === 'tanh') { yMin = -0.02; yMax = 1.02; yTicks = [0,0.25,0.5,0.75,1]; }
        if (state.activation === 'leaky') { yMin = -0.02; yMax = 1.08; yTicks = [0,0.1,0.5,1]; }

        const maps = drawAxes(ctx, w, h, xMin, xMax, yMin, yMax, [-6,-4,-2,0,2,4,6], yTicks, area);
        const { xMap, yMap } = maps;
        drawRegionBands(ctx, xMap, area.mt, h - area.mb);

        ctx.strokeStyle = "#b86a1a";
        ctx.lineWidth = 3;
        ctx.beginPath();
        for (let i = 0; i <= 320; i++) {
          const x = xMin + (xMax - xMin) * i / 320;
          const y = df(x);
          const px = xMap(x);
          const py = yMap(y);
          if (i === 0) ctx.moveTo(px, py);
          else ctx.lineTo(px, py);
        }
        ctx.stroke();

        const cx = xMap(state.z);
        const cy = yMap(df(state.z));

        ctx.setLineDash([5, 6]);
        ctx.strokeStyle = "rgba(184,106,26,0.22)";
        ctx.lineWidth = 1.6;
        ctx.beginPath();
        ctx.moveTo(cx, yMap(yMin));
        ctx.lineTo(cx, cy);
        ctx.stroke();
        ctx.beginPath();
        ctx.moveTo(xMap(0), cy);
        ctx.lineTo(cx, cy);
        ctx.stroke();
        ctx.setLineDash([]);

        ctx.fillStyle = "#ffffff";
        ctx.strokeStyle = "#b86a1a";
        ctx.lineWidth = 3;
        ctx.beginPath();
        ctx.arc(cx, cy, 6.2, 0, Math.PI * 2);
        ctx.fill();
        ctx.stroke();

        ctx.fillStyle = "#b86a1a";
        ctx.font = "700 12px Inter, sans-serif";
        ctx.textAlign = "left";
        ctx.fillText(`g'(z) = ${fmt(df(state.z))}`, cx + 12, cy - 10);
      }

      function updateText() {
        const meta = META[state.activation];
        const region = regionInfo(state.z);

        badge.textContent = meta.name;
        regionBadge.textContent = region.short;
        zVal.textContent = fmt(state.z);
        outVal.textContent = fmt(f(state.z));
        dfVal.textContent = fmt(df(state.z));

        formula.innerHTML =
`<strong>${meta.name}</strong>
${meta.formula}

<strong>Derivative</strong>
${meta.derivative}`;

        explainTitle.textContent = meta.title;
        explainText.textContent = meta.text;

        statusText.textContent = region.label;
        statusDot.style.background = region.color;
        status.style.color = region.color;
        status.style.background = region.bg;
        status.style.borderColor = region.color + "33";

        document.querySelectorAll('#fix_act_buttons .btn').forEach(btn => {
          btn.classList.toggle('active', btn.dataset.act === state.activation);
        });
      }

      function render() {
        updateText();
        drawFunctionChart();
        drawDerivativeChart();
      }

      function loop() {
        if (!state.playing) return;
        state.z += 0.04 * state.dir;
        if (state.z >= 6) { state.z = 6; state.dir = -1; }
        if (state.z <= -6) { state.z = -6; state.dir = 1; }
        zSlider.value = state.z;
        render();
        state.raf = requestAnimationFrame(loop);
      }

      document.querySelectorAll('#fix_act_buttons .btn').forEach(btn => {
        btn.addEventListener('click', () => {
          state.activation = btn.dataset.act;
          render();
        });
      });

      zSlider.addEventListener('input', e => {
        state.z = parseFloat(e.target.value);
        render();
      });

      playBtn.addEventListener('click', () => {
        if (state.playing) return;
        state.playing = true;
        state.raf = requestAnimationFrame(loop);
      });

      pauseBtn.addEventListener('click', () => {
        state.playing = false;
        if (state.raf) cancelAnimationFrame(state.raf);
      });

      resetBtn.addEventListener('click', () => {
        state.playing = false;
        if (state.raf) cancelAnimationFrame(state.raf);
        state.activation = 'relu';
        state.z = 1.2;
        state.dir = 1;
        zSlider.value = state.z;
        render();
      });

      window.addEventListener('resize', render);
      setTimeout(render, 80);
      render();
    })();
  </script>
</div>
"""))
```
